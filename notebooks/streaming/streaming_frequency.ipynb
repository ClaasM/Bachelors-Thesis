```python
# Data collected by printing to the console on the dashboard
sample = [10,43,8,17,18,30,12,7,7,24]
filter = [3,2,1,2,0,2,3,7,3,5]

import matplotlib.pyplot as plt

plt.plot(sample, label="sample stream")
plt.plot(filter, label="filter stream")
for a,b in enumerate(sample): 
    plt.text(a, b, str(b))
for a,b in enumerate(filter): 
    plt.text(a, b, str(b))
plt.legend()
plt.savefig("stream_frequencies.pdf", type="pdf")
plt.show()
```
