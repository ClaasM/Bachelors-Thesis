```python
from pyspark.streaming import StreamingContext
from pyspark import SparkContext

import time
```

```python
sc = SparkContext.getOrCreate()
ssc = StreamingContext(sc, 1)
#Create the queue through which RDDs can be pushed to a QueueInputDStream

rddQueue = [sc.emptyRDD()]
# Create the QueueInputDStream and use it do some processing
ssc.queueStream(rddQueue).pprint() # <- This doesnt print anything because the queue is filled after stream creation
ssc.start()

# Create and push some RDDs into
for i in range(30):
  rddQueue += [ssc.sparkContext.parallelize(range(10), 5)]
  time.sleep(1)

ssc.stop()
```

Adding to the rddQueue after the Stream is created is only supported in the Scala API, so this doesn't work.
[See Documentation for queueStream](https://spark.apache.org/docs/2.2.0/api/python/pyspark.streaming.html#pyspark.streaming.StreamingContext)
